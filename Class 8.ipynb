# Class 8: Version Control

### Getting Started with Github

#### Why use Github at all?

Github helps keep the most recent files updated for all members of a team. It's to avoid a situation where we are sending files to each other with "finalfinal2.csv" and so forth. Multiple people can work on the same project using their own files for new features, fixing bugs etc.
