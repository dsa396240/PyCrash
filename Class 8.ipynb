# Class 8: Version Control

### Getting Started with Github

#### Why use Github at all?

Github helps keep the most recent files updated for all members of a team. It's to avoid a situation where we are sending files to each other with "finalfinal2.csv" and so forth. Multiple people can work on the same project using their own files for new features, fixing bugs etc. 

#### Difference Between Git and Github

Git is an open-source, version control tool created in 2005 by developers working on the Linux operating system; GitHub is a company founded in 2008 that makes tools which integrate with git. You do not need GitHub to use git, but you cannot use GitHub without using git.

### You can create an SSH connection with GitHub to allow seamless integration. What is an SSH key?

A Secure Shell (SSH) protocol uses a pair of keys to initiate a secure handshake between remote parties. The key pair contains a public and private key. The private vs public nomenclature can be confusing as they are both called keys. It is more helpful to think of the public key as a "lock" and the private key as the "key". You give the public 'lock' to remote parties to encrypt or 'lock' data. This data is then opened with the 'private' key which you hold in a secure place. 

### Getting familiar with Git Commands

#### We connect to a remote Git Repo using:

"git remote add origin git@github.com:username/repo.git"

## Commands:

### On your local folder, you will "initialize" a new Git Repo and connect it to the remote Git Repo on Github.

"git init" - Create the local git repository.

### This one is important, you'd like to check often whether a file is loaded to the commit stage or not.

"git status' - Check the condition of files and any changes made within them.

### Check connection to the Remote Git Server on GitHub

"git remote -v" check out remote git server

### The staging area holds all the files that will be sent to the remote repo on a "push"

"git add [filename]" - Add the file to the git staging area

### Remember to post meaningful comments so your teammates know what changes have been made

"git commit -m "message"" - Commit the file and add a message to track what changes have been made

### Ship your latest code to the remote repo

"git push" - Push the file to the remote Git Repository
